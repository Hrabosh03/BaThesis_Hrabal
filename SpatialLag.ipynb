```python
import geopandas as gpd
import libpysal
import numpy as np
import os

# --- Vstupy ---
vstup_shp = r"C:\Skola\BP\data_analyza\skoly_polygony\skoly2017_polygony.shp"
swm_path = r"C:\Skola\BP\vizualizace\matice_skoly\skoly2017.swm"
vystup_shp = r"C:\Skola\BP\vrstvy_app_final\SL_zaci\zaci2017SL.shp"
sloupec = "Index" 

# Naten dat
gdf = gpd.read_file(vstup_shp)
x = gdf[sloupec].astype(float).values

# Naten vhov matice
f = libpysal.io.open(swm_path)
w = f.read()
f.close()
w.transform = "r"

# Vpoet spatial lag
spatial_lag = w.sparse @ x
gdf["spatial_lag"] = spatial_lag  # spatial lag = ven prmr hodnot sousednch zsj

# Uloen s kdovnm UTF-8
if os.path.exists(vystup_shp):
    os.remove(vystup_shp)

gdf.to_file(vystup_shp, encoding="utf-8")

print("Spatial lag spn spotn a uloen do shapefilu (UTF-8).")
```
